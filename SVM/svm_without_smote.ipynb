```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
train_df = pd.read_csv("train.csv") #loading the training data
```

```python
train_df.drop_duplicates(inplace=True) #Removing duplicates
```

```python
#dropping unnecessary columns
train_df.drop('LoanID', axis=1, inplace=True)
```

```python
print(len(pd.unique(train_df['Education'])))
print(len(pd.unique(train_df['EmploymentType'])))
print(len(pd.unique(train_df['MaritalStatus'])))
print(len(pd.unique(train_df['LoanPurpose'])))
```

```python
#Label encoding categorical columns
le = LabelEncoder()
for col in ['Education', 'EmploymentType', 'MaritalStatus', 'LoanPurpose', 'HasMortgage', 'HasDependents', 'HasCoSigner']:
    train_df[col] = le.fit_transform(train_df[col])
```

```python
train_df.head()
```

```python
X = train_df.drop('Default', axis=1)
y = train_df['Default']

# Split data after preprocessing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#training the svm model
svc=SVC()

print("Default Parameters of SVC:")
print(svc.get_params())

svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print('Model accuracy score with default hyperparameters: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
import pickle
#Loading test and sample_submission files
test_df = pd.read_csv("test.csv")
sample_submission = pd.read_csv("sample_submission.csv")

#Same pre-processing steps for the testing 
test_df.drop('LoanID', axis=1, inplace=True)
le = LabelEncoder()
for col in ['Education', 'EmploymentType', 'MaritalStatus', 'LoanPurpose', 'HasMortgage', 'HasDependents', 'HasCoSigner']:
    test_df[col] = le.fit_transform(test_df[col])

X_test = test_df 
predictions = svc.predict(X_test)

submission_df = sample_submission.copy()
submission_df['Default'] = predictions

submission_df.to_csv("submission_svm_default_without_smote.csv", index=False)
print("submission_svm_default.csv file created successfully.")
```

```python
#using GridSearchCV for finding the best parameters
from sklearn.model_selection import GridSearchCV
svc = SVC()

param_grid = {
    'C': [0.1, 1, 10, 100],  #Regularization parameter
    'gamma': [1, 0.1, 0.01, 0.001],  #Kernel coefficient for 'rbf'
    'kernel': ['rbf', 'linear']  #Kernel type
}

grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, scoring='accuracy', cv=5, verbose=2)
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Score:", grid_search.best_score_)
```
